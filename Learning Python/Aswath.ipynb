```python
def square(array):
    n = len(array)
    temp = []
    for i in range (n):
        temp.append(array[i] ** 2)
    return temp
```

```python
def squareroot(array):
    n = len(array)
    temp = []
    for i in range (n):
        temp.append(array[i] ** 0.5)
    return temp
```

```python
inputValue = []
for i in range (1,101):
    inputValue.append(i)
squareList = square(inputValue)
print(squareList, sep = ", ")
```

```python
squarerootList = squareroot(squareList)
print(squarerootList, sep = ", ")
```
